## Retrieving the Data of fog-harbor-fish-house

```python
import pickle
import requests
from bs4 import BeautifulSoup 
URL= 'https://www.yelp.com/biz/fog-harbor-fish-house-san-francisco-2'
page=requests.get(URL)
soup = BeautifulSoup(page.text, 'html5lib')
rest_review1 = soup.find_all('p', {'lang':'en'})
restaurant_list = ['Fog Harbor Fish House', 'The House', 'Marufuku Ramen SF', 'Zazie', 'Kitchen Story']
rest_review1_list = list()
string = ""
for i in rest_review1:
    string += i.text
    #print(i.text,'\n')
    rest_review1_list.append(i.text)

print(rest_review1_list)
```

## Sentiment Analysis for fog-harbor-fish-house

        Determines the positive, negative, neutral and overall compound sentiment for the review 


```python
import pickle
from nltk_helpers import get_sentiments

for empty in rest_review1_list:
    print(get_sentiments(empty))
```

## Sentiment Analysis graph for fog-harbor-fish-house

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

#add another set of data
import pickle

from nltk_helpers import split_sentiments

arbyX, arbyY, arbyZ = split_sentiments([get_sentiments(empty) for empty in rest_review1_list])
ax.scatter(arbyX, arbyY, arbyZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_zlabel('Neutral')
ax.set_ylabel('Positive')
ax.set_title('Sentimental Analysis Plot for fog-harbour-fish-house')
plt.show()
```

## Retrieving the Data of The House Restaurant

```python
import pickle
import requests
from bs4 import BeautifulSoup 
URL1= 'https://www.yelp.com/biz/the-house-san-francisco'
page1=requests.get(URL1)
soup1 = BeautifulSoup(page1.text, 'html5lib')
rest_review2 = soup1.find_all('span', {'lang':'en','class':'lemon--span__373c0__3997G'})
#print(rest_review2)
rest_review2_list = list()
for i in rest_review2:
    string += i.text
    #print(i.text,'\n')
    rest_review2_list.append(i.text)
print(rest_review2_list)
```

## Sentiment Analysis for The House Restaurant

```python
import pickle
from nltk_helpers import get_sentiments

for empty in rest_review2_list:
    print(get_sentiments(empty))
```

## Sentimental Analysis graph for The House

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

#add another set of data
import pickle

from nltk_helpers import split_sentiments

arbyX, arbyY, arbyZ = split_sentiments([get_sentiments(empty) for empty in rest_review2_list])
ax.scatter(arbyX, arbyY, arbyZ, color='b', marker='^')

ax.set_xlabel('Negative')
ax.set_zlabel('Neutral')
ax.set_ylabel('Positive')
ax.set_title('Sentimental Analysis Plot for The House')

plt.show()
```

## Retrieving the Data of  Marufuku-Ramen-sf Restaurant

## Complete Sentiment Analysis graph for all reviews 

```python
import matplotlib.pyplot as plt
import matplotlib.style as style

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

#add another set of data

from nltk_helpers import split_sentiments
style.use('default')
arbyX, arbyY, arbyZ = split_sentiments([get_sentiments(empty) for empty in rest_review1_list])
ax.scatter(arbyX, arbyY, arbyZ, color='r', marker='^', label='fog-harbor-fish-house')

arbX, arbY, arbZ = split_sentiments([get_sentiments(empty) for empty in rest_review2_list])
ax.scatter(arbX, arbY, arbZ, color='b', marker='^', label='the-house')

arX, arY, arZ = split_sentiments([get_sentiments(empty) for empty in rest_review3_list])
ax.scatter(arX, arY, arZ, color='g', marker='^', label='marufuku-ramen-sf')

ax.set_xlabel('Negative')
ax.set_zlabel('Neutral')
ax.set_ylabel('Positive')
ax.legend()
ax.set_title('Sentimental Analysis Plot')


plt.show()
```
